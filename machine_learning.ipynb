3. For each of the following real-world problems, determine whether it is appropriate to utilise a supervised or unsupervised machine learning algorithm. Justify your answers:
* 3.1 Detecting anomalies in a manufacturing process using sensor data without prior knowledge of specific anomaly patterns.
* 3.2 Predicting customer lifetime value based on historical transaction data and customer demographics.
* 3.3 Segmenting customer demographics based on their purchase history, browsing behaviour, and preferences.
* 3.4 Analysing social media posts to categorise them into different themes.


3. Answer here
* 3.1 Appropriate ML algorithm: Unsupervised learning; as there is no labelled data for prior specific anomaly patterns. This unsupervised learning approach can identify outliers.
* 3.2 Appropriate ML algorithm: Supervised learning; a regression problem predicting the target variable ‘customer lifetime value’ using labelled historical transaction and customer data with known outcomes.
* 3.3 Appropriate ML algorithm: Unsupervised learning; identifying patterns in the data without predefined classes or labelled outcomes.
* 3.4 Appropriate ML algorithm: Unsupervised or supervised depending on the required task outcome when analysing social media post categorisation. Unsupervised learning is appropriate for initial clustering. However, when it is necessary to be more specific about a theme categorisation, supervised learning based on labelled data is more appropriate.

For each of the following real-world problems, determine whether it is appropriate to utilise semi-supervised machine learning algorithms. Justify your answers:
* 4.1 Predicting fraudulent financial transactions using a dataset where most transactions are labelled as fraudulent or legitimate.
* 4.2 Analysing customer satisfaction surveys where only a small portion of the data is labelled with satisfaction ratings.
* 4.3 Identifying spam emails in a dataset where the majority of emails are labelled.
* 4.4 Predicting the probability of default for credit card applicants based on their complete financial and credit-related information.


4. Answer here
* 4.1 Inappropriate to utilise semi-supervised ML algorithms: a significant amount of labelled data is available. In this scenario it is appropriate to use supervised learning where there is already clear labelling of fraud or legitimate transactions.
* 4.2 Appropriate to utilise semi-supervised ML algorithms: the ML model performance can be improved using semi-supervised learning, making use of a combination of both labelled and unlabelled data from the customer satisfaction surveys. This is particularly useful in the scenario given where there is a small limited portion of labelled data with satisfaction ratings and a large unlabelled dataset that is missing this information.
* 4.3 Inappropriate to utilise semi-supervised ML algorithms: as there is already a sufficient amount of labelled email data available for supervised learning. In this scenario there is already clear labelling of spam and non-spam emails and as such makes supervised learning more appropriate.
* 4.4 Inappropriate to utilise semi-supervised ML algorithms: as applicants have complete financial and credit-related labelled information recorded in the dataset it would be suitable to approach this problem with a supervised learning model. There is no suggestion that only a small amount of labelled data is available, and as such a purely supervised approach is more suitable for predicting credit card default probabilities of applicants based on the available complete financial and credit-related data.
